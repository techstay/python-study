```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data for the sine wave
x = np.linspace(0, 2 * np.pi, 500)
y = np.sin(x)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the sine wave
ax.plot(x, y)

# Set the x and y axis labels
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")

# Set the plot title
ax.set_title("Sine Wave")

# Show the plot
plt.show()
```
